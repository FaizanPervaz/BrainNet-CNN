```python
import os
import nibabel as nib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tqdm import tqdm

def load_nifti_slice(nifti_path):
    img = nib.load(nifti_path)
    data = img.get_fdata()
    mid_slice = data.shape[2] // 2  # Take the middle slice
    image = data[:, :, mid_slice]
    image = (image - np.min(image)) / (np.max(image) - np.min(image))  # Normalize
    return image

def get_brats_images(root_dir, label):
    image_list = []
    for folder in tqdm(os.listdir(root_dir)):
        folder_path = os.path.join(root_dir, folder)
        t1ce_file = os.path.join(folder_path, f"{folder}_t1ce.nii")
        if os.path.exists(t1ce_file):
            try:
                img = load_nifti_slice(t1ce_file)
                image_list.append((img, label))
            except Exception as e:
                print(f"Error reading {t1ce_file}: {e}")
    return image_list

# Define paths
hgg_path = r"C:\Users\Faizy\Downloads\dataset\MICCAI_BraTS_2019_Data_Training\HGG"
lgg_path = r"C:\Users\Faizy\Downloads\dataset\MICCAI_BraTS_2019_Data_Training\LGG"

# Load Data
hgg_images = get_brats_images(hgg_path, "HGG")
lgg_images = get_brats_images(lgg_path, "LGG")

# Combine
all_brats_data = hgg_images + lgg_images

# Example visualization
plt.imshow(all_brats_data[0][0], cmap='gray')
plt.title(all_brats_data[0][1])
plt.show()
```

```python
import numpy as np
import random

#   Plotting
import seaborn as sns
import matplotlib.pyplot as plt


#   DataGenerator to read images and rescale images
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import tensorflow as tf
import tensorflow_addons as tfa

#   count each class samples
from collections import Counter

#   callbacks
from tensorflow.keras.callbacks import ReduceLROnPlateau

#   evaluate precison recall and f1-score of each class of model
from sklearn.metrics import classification_report
#   Show performance of a classification model
from sklearn.metrics import confusion_matrix

#   Different layers
from keras.models import Sequential
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import MaxPool2D
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import ReLU
from tensorflow.keras.layers import Softmax

# split dataset to train, validation and test set
from sklearn.model_selection import train_test_split

#   callbacks
from keras import callbacks

#   ADASYN from imblance library
from imblearn.over_sampling import ADASYN

#   Optimizer
from tensorflow.keras.optimizers import SGD

from sklearn.metrics import roc_curve, auc
from itertools import cycle
```

```python
X, y = [], []

datagen = ImageDataGenerator(rescale=1.0/255.0)

# Load images by resizing and shuffling randomly
train_dataset = datagen.flow_from_directory(
    WORKING_DIRECTORY,
    target_size=(IMG_SIZE, IMG_SIZE),  # Resize images to 150x150
    batch_size=6400,  # Change batch size to load images at once
    class_mode='categorical',  # Since you have multiple classes
    shuffle=True  # Shuffle the images randomly
)

# Separate dataset from Data Generator
X, y = train_dataset.next()
```

Implementation of ADASYN

```python
# Reshape the images to 1D for ADASYN
X_flat = X.reshape(-1, IMG_SIZE * IMG_SIZE * 3)

# Apply ADASYN to balance the classes
X_resampled, y_resampled = ADASYN().fit_resample(X_flat, y)

# Reshape back to image dimensions
X = X_resampled.reshape(-1, IMG_SIZE, IMG_SIZE, 3)
y = y_resampled  # Make sure to update y as well

# Count the number of samples per class after oversampling
samples_after = len(X)
print("Number of samples after ADASYN:\t", sorted(Counter(np.argmax(y, axis=1)).items()))
```

CNN Implementation

```python
from keras.initializers import LecunUniform
init = LecunUniform

model = Sequential()

model.add(Input(shape=(IMG_SIZE, IMG_SIZE, 3)))

model.add(Conv2D(8, 3, padding="same", kernel_initializer=init))
model.add(ReLU())
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(16, 3, padding="same", kernel_initializer=init))
model.add(ReLU())
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(32, 3, padding="same", kernel_initializer=init))
model.add(ReLU())
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(64, 3, padding="same", kernel_initializer=init))
model.add(ReLU())
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(128, 3, padding="same", kernel_initializer=init))
model.add(ReLU())
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(256, 3, padding="same", kernel_initializer=init))
model.add(ReLU())
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(512, 3, padding="same", kernel_initializer=init))
model.add(ReLU())
model.add(MaxPool2D(pool_size=(2,2)))


model.add(Dropout(0.1))

model.add(Flatten())

model.add(Dense(512, kernel_initializer=init))
model.add(ReLU())

model.add(Dense(3, kernel_initializer=init))
model.add(Softmax())

model.summary()
```

```python
import tensorflow_addons as tfa
from tensorflow.keras.optimizers import RMSprop

model.compile(
    optimizer=RMSprop(learning_rate=0.0001),
    loss=tf.keras.losses.CategoricalCrossentropy(name='loss'),
    metrics=[
        tf.keras.metrics.CategoricalAccuracy(name='acc'),
        tf.keras.metrics.AUC(name='auc'),
        tfa.metrics.F1Score(num_classes=3, average='macro', name='f1_score'),
        tf.keras.metrics.Precision(name="precision"),
        tf.keras.metrics.Recall(name="recall")
    ]
)
```

```python
# callbacks used in model to perform well
rop_callback = ReduceLROnPlateau(monitor="val_loss")

CALLBACKS = [rop_callback]
```

```python
#   declare to run on small gpu create batch sizes of images
valAug = ImageDataGenerator()

#   defining batch size
batch_size = 32

history = model.fit(valAug.flow(X_train, y_train, batch_size=batch_size),
steps_per_epoch=len(X_train) // batch_size,
validation_data=valAug.flow(x_val, y_val, shuffle = True),
validation_steps=len(x_test) // batch_size,
epochs= 14,
batch_size=batch_size,
callbacks = CALLBACKS
)
```

```python
test_scores = model.evaluate(x_test, y_test, batch_size=32)

# Print the results
print("\n\nTesting Loss : \t\t {0:0.6f}".format(test_scores[0]))
print("Testing Accuracy : \t {0:0.6f} %".format(test_scores[1] * 100))
print("Testing AUC : \t\t {0:0.6f} %".format(test_scores[2] * 100))  # AUC
print("Testing F1-Score : \t {0:0.6f} %".format(test_scores[3] * 100))  # F1-score
print("Testing Precision : \t {0:0.6f} %".format(test_scores[4] * 100))  # Precision
print("Testing Recall : \t {0:0.6f} %".format(test_scores[5] * 100))  # Recall
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import numpy as np
import matplotlib.pyplot as plt

# Ensure predictions are probability scores
y_score = model.predict(x_test)  # shape: (num_samples, 3)
y_test_bin = y_test              # assuming already one-hot encoded

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
n_classes = 3

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Compute micro-average ROC curve and ROC area
fpr["micro"], tpr["micro"], _ = roc_curve(y_test_bin.ravel(), y_score.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# Compute macro-average ROC curve and ROC area
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))

mean_tpr = np.zeros_like(all_fpr)
for i in range(n_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])

mean_tpr /= n_classes

fpr["macro"] = all_fpr
tpr["macro"] = mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

# Plot all ROC curves
plt.figure()
ax = plt.axes()
ax.set_facecolor('#EBEBEB')

plt.plot(fpr["micro"], tpr["micro"],
         label="micro-average ROC curve (area = {0:0.4f})".format(roc_auc["micro"]),
         color="deeppink", linestyle=":", linewidth=4)

plt.plot(fpr["macro"], tpr["macro"],
         label="macro-average ROC curve (area = {0:0.4f})".format(roc_auc["macro"]),
         color="navy", linestyle=":", linewidth=4)

colors = ['red', 'green', 'blue']
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], color=colors[i], lw=2,
             label="ROC curve of class {0} (area = {1:0.4f})".format(i, roc_auc[i]))

plt.plot([0, 1], [0, 1], "k--", lw=2)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Multiclass ROC Curve")
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

Predictions with Heat Cam 

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
from tensorflow.keras.preprocessing import image


def grad_cam(model, img_array, class_index=None):
    """
    Generate the Grad-CAM heatmap for the model's predictions.
    
    Parameters:
        model: Keras model
        img_array: Input image (processed and reshaped)
        class_index: The class to visualize, leave as None to use predicted class.
        
    Returns:
        heatmap: The Grad-CAM heatmap.
    """
    # Get the last convolutional layer
    last_conv_layer = model.get_layer('conv2d_7')  # Change to your last conv layer name
    last_conv_model = tf.keras.models.Model(inputs=model.input, outputs=last_conv_layer.output)
    
    # Model for the gradient of the class output w.r.t. the last conv layer's output
    grad_model = tf.keras.models.Model(inputs=model.input, outputs=[last_conv_layer.output, model.output])
    
    # Get model predictions
    with tf.GradientTape() as tape:
        conv_output, predictions = grad_model(img_array)
        if class_index is None:
            class_index = np.argmax(predictions[0])  # Use the predicted class if none provided
        
        # Get the output for the selected class
        class_output = predictions[:, class_index]
    
    # Get the gradients of the class output w.r.t. the last conv layer's output
    grads = tape.gradient(class_output, conv_output)
    
    # Global average pooling
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    
    # Multiply the pooled gradients by the output feature map
    conv_output = conv_output[0]
    heatmap = np.dot(conv_output, pooled_grads.numpy())
    
    # Apply ReLU (to focus on positive influences)
    heatmap = np.maximum(heatmap, 0)
    
    # Normalize the heatmap
    heatmap /= np.max(heatmap)
    
    return heatmap

def overlay_heatmap_on_image(img_path, heatmap):
    """
    Overlay the heatmap on the original image for visualization.
    
    Parameters:
        img_path: Path to the original image.
        heatmap: Grad-CAM heatmap.
        
    Returns:
        output_image: Image with the heatmap overlayed.
    """
    # Load the image
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    
    # Resize heatmap to match the input image dimensions
    heatmap_resized = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
    
    # Convert the heatmap to a color map
    heatmap_color = cv2.applyColorMap(np.uint8(255 * heatmap_resized), cv2.COLORMAP_JET)
    
    # Overlay the heatmap on the original image
    output_image = np.uint8(0.5 * heatmap_color + 0.5 * img)
    
    return output_image

def generate_grad_cam_for_image(image_path, model, class_index=None):
    """
    Generate Grad-CAM for a single image and display the result.
    
    Parameters:
        image_path: Path to the image for which Grad-CAM is generated.
        model: The trained model.
        class_index: Index of the class to visualize (optional).
    """
    # Load the image
    img = image.load_img(image_path, target_size=(150, 150))  # Resize to input shape
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    
    # Normalize the image
    img_array = img_array / 255.0  # We Assume the model expects images scaled from 0 to 1
    
    # Generate the Grad-CAM heatmap
    heatmap = grad_cam(model, img_array, class_index)
    
    # Overlay the heatmap on the image
    output_image = overlay_heatmap_on_image(image_path, heatmap)
    
    # Displaying the original image and the image with the heatmap overlay
    plt.figure(figsize=(10, 10))
    plt.subplot(1, 2, 1)
    plt.imshow(img_array[0])
    plt.title("Original Image")
    
    plt.subplot(1, 2, 2)
    plt.imshow(output_image)
    plt.title("Grad-CAM Heatmap Overlay")
    
    plt.show()

# Example usage
image_path = r"C:\Users\Faizy\Downloads\dataset\Training\Healthy\IXI002-Guys-0828-T1_70.png" 

generate_grad_cam_for_image(image_path, model)
```
